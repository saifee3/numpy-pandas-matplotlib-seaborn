```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from bokeh.plotting import figure,output_file,show
iris=sns.load_dataset('iris')
```

# Complex Plot

```python
sns.set()
A = np.linspace(0,100,1000)
B = np.cumsum(np.random.randn(1000,5),0)
plt.plot(A,B)
plt.legend('ABCDE',ncol=2,loc='upper left')
```

## Simple Plot

```python
df =pd.DataFrame({'X':np.random.normal(0,2,1000),
                  'Y':np.random.normal(3,5,1000),
                  'Z':np.random.normal(-4,10,1000)
                 })
```

```python
df.head()
```

```python
sns.set()
plt.figure(figsize=(4,3))
for col in df.columns:
    plt.hist(df[col],density=True,alpha=0.5,bins=30)
    sns.kdeplot(df[col],shade=True)
```

## Distplot

```python
df =pd.DataFrame({'X':np.random.normal(0,2,1000),
                  'Y':np.random.normal(3,5,1000),
                  'Z':np.random.normal(-4,10,1000)
                 })
plt.figure(figsize=(4,3))
for col in df.columns:
    sns.distplot(df[col])
```

## Jointplot

```python
df =pd.DataFrame({'X':np.random.normal(0,2,1000),
                  'Y':np.random.normal(3,5,1000),
                  'Z':np.random.normal(-4,10,1000)
                 })
with sns.axes_style('white'):
    sns.jointplot('X','Y',df,kind='kde')
```

## Relation Plot

```python
plt.figure(figsize=(4,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species')
```

### Line Kind

```python
plt.figure(figsize=(4,3))
sns.relplot(x='sepal_length',y='sepal_width',dashes=True,markers=True,data=iris,kind='line',ci='sd',hue='species',style='species')
# ci can be None 
# play with hue and style too
```
