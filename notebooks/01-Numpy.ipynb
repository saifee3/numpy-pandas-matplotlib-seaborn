### Breaking into submatrices 

```python
import numpy as np

# Create a 16x16 array
A = np.arange(64).reshape(8, 8)

# Separate the array into 4x4 submatrices
sub_matrices = np.vsplit(A, 4)
sub_matrices = [np.hsplit(sub_matrix, 4) for sub_matrix in sub_matrices]
sub_matrices = np.array(sub_matrices).reshape(4,4,4)
```

## K-Means Clustering (KNN) from scratch using Numpy (Not Complete Lec 108) 

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
print("Shape of Iris Dataset:",iris.data.shape,"\n")
print(iris.data)
```

```python
X = iris.data
Y = iris.target
```

```python
def myTrainTestSplit(X,Y,test_size):
    idx = np.random.permutation(np.arange(X.shape[0]))
    tS = int(np.floor(test_size*X.shape[0]))
    testIdx = idx[:tS]
    trainIdx = idx[tS:]
                   
    Xtrain = X[trainIdx,:]
    Ytrain = Y[trainIdx]
    Xtest = X[testIdx,:]               
    Ytest = Y[testIdx]                   
    return Xtrain,Xtest,Ytrain,Ytest
```

```python
Xtrain,Xtest,Ytrain,Ytest= myTrainTestSplit(X,Y,0.2)
```

```python
D=np.sum((Xtrain[:,np.newaxis,:]-Xtest[np.newaxis,:,:])**2,axis=2)
```

```python
nidx=np.argsort(D,axis=1)
```
